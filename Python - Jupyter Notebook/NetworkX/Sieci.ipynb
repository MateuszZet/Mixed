```python
import networkx as nx
import matplotlib.pyplot as plt
import random as random
from tqdm import tqdm
```

```python
g=nx.Graph()
for x in range(0,14):
    g.add_edge(x,random.randint(0,14),weight=random.randint(1,50)),


nx.draw_spring(g, with_labels=True)
plt.show()
```

```python
g=nx.barabasi_albert_graph(8,2)
nx.draw_spring(g, with_labels=True)
plt.show()
```

```python
stat = nx.info(g)
print(stat)
print(g.edges)
```

```python
for x in g.nodes:
    t = nx.bfs_tree(g,x)
    nx.draw_spring(t,with_labels=True)
    plt.show()
    print(list(nx.bfs_edges(g,x))),
```

```python
for x in g.nodes():
    for y in g.nodes():
        print(list(nx.shortest_path(g,x,y)), " ", nx.shortest_path_length(g,x,y)),
```
